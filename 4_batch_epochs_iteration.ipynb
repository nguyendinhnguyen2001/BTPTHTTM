## epoch là gì?
-là tổng số lần lặp lại để đào tạo mô hình học máy với tất cả dữ liệu đào tạo trong một chu kỳ. Trong Epoch, tất cả dữ liệu đào tạo được sử dụng chính xác một lần. Nói cách khác, Epoch còn có thể được hiểu là tổng số lần vượt qua mà một thuật toán đã hoàn thành xung quanh tập dữ liệu huấn luyện.

## iteration là gì?
-Lặp lại được định nghĩa là tổng số lô được yêu cầu để hoàn thành một kỷ nguyên, trong đó số lô bằng tổng số lần lặp lại cho một kỷ nguyên.

## batch là gì?
-Kích thước lô là một siêu thông số xác định số lượng mẫu cần xử lý trước khi cập nhật các thông số mô hình bên trong.
