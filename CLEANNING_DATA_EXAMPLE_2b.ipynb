```python
pub = df['Place of Publication']#khởi tạo pub là dữ liệu cột Place of Publication
london = pub.str.contains('London')#khởi tạo mảng london là mảng kiểu boolean kt chuỗi London có trong phần tử của pub ko?
london[:10]
```

```python
oxford = pub.str.contains('Oxford')#khởi tạo mảng london là mảng kiểu boolean kt chuỗi Oxford có trong phần tử của pub ko?
```

```python
#thay thế các dữ liệu trong df['Place of Publication'] chứa Oxford-->Oxford,chứa London-->London
df['Place of Publication'] = np.where(london, 'London',np.where(oxford, 'Oxford',pub.str.replace('-', ' ')))
```

```python
#khởi tạo mảng university_towns
university_towns = []
#đọc file university_towns.txt
with open('university_towns.txt') as file:
    for line in file:
        if '[edit]' in line:
            state = line#state chứa '[edit]'
        else:
            university_towns.append((state, line))#thêm vào university_towns cặp(state,line)
university_towns[:5]#hiện 5 phần tử đầu của mảng university_towns
```

```python
towns_df = pd.DataFrame(university_towns,columns=['State', 'RegionName'])#khởi tạo towns_df là dataFrame của mảng university_towns 
```

```python
towns_df.head()#hiện 5 phần tử đầu của towns_df
```

```python
#hàm xử lý dữ liệu
def get_citystate(item):
    if ' (' in item:#check '(' trong item,nếu có thì lấy cắt các giá trị trước '('
        return item[:item.find(' (')]
    elif '[' in item:#check '[' trong item,nếu có thì lấy cắt các giá trị trước '['
        return item[:item.find('[')]
    else:#nếu không có thì trả về item
        return item
```

```python
towns_df =  towns_df.applymap(get_citystate)#sử dụng hàm get_citystate để xử lý dữ liệu trong towns_df
```
