# Cc hm kch hot (activation function) trong neural network

Hm kch hot (activation function) m phng t l truyn xung qua axon ca mt neuron thn kinh. Trong mt mng n-ron nhn to, hm kch hot ng vai tr l thnh phn phi tuyn ti output ca cc n-ron.
